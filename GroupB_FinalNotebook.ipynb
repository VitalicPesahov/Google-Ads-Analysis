**Find the top 10 keywords based on Cost per Impression, Clicks, and Cost per Click**






```python
# convert cost column from string to float
cost_list=keyword_data['Cost'].to_list()

cost_float_list=[]
for c in cost_list:
  cost_float_list.append(float(c))

keyword_data['Cost']= cost_float_list

# convert interactions column from string to float
interactions_list=keyword_data['Interactions'].to_list()

interactions_float_list=[]
for c in interactions_list:
  interactions_float_list.append(float(c))

keyword_data['Interactions']= interactions_float_list

# convert impression column from string to float
impression_list=keyword_data['Impr.'].to_list()
impression_float_list=[]
import locale
locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')
for i in impression_list:
  i =locale.atof(i)
  impression_float_list.append(i)

keyword_data['Impr.']= impression_float_list
```

```python
# solve for Cost per Impression and Cost per Interation, add them into dataframe
keyword_data['Cost per Impression'] = keyword_data['Cost']/keyword_data['Impr.']
keyword_data['Cost per Interaction'] = keyword_data['Cost']/keyword_data['Interactions']
keyword_data
```

```python
# show the top 10 keywords based on Cost Per Impression
top10_keywords_cost_per_impr = keyword_data.nlargest(10, 'Cost per Impression')
top10_keywords_cost_per_impr
```

```python
# show the top 10 keywords based on Interactions (clicks)
top10_keywords_interactions = keyword_data.nlargest(10, 'Interactions')
top10_keywords_interactions
```

```python
# show the top 10 keywords based on Cost per Interaction
top10_keywords_cost_per_interaction = keyword_data.nlargest(10, 'Cost per Interaction')
top10_keywords_cost_per_interaction
```

```python
# find the common top 10 keywords among Cost per Impression, Clicks, and Cost per Click
cost_per_impr_list=top10_keywords_cost_per_impr.index.to_list()
interactions_list=top10_keywords_interactions.index.to_list()
cost_per_interaction_list=top10_keywords_cost_per_interaction.index.to_list()

top10_common_keywords = list(set.intersection(*map(set, [cost_per_impr_list, interactions_list, cost_per_interaction_list])))
if not top10_common_keywords:
  print("No common top 10 keywords among Cost per Impression, Clicks, and Cost per Click")
else:
  print(top10_common_keywords)
```

**Find the keywords which have lower Average Cost (Cost per Interaction) and higher Interaction Rate**

```python
# show the bottom 15 keywords based on Average Cost (Cost per Interaction)
bottom15_keywords_cost_per_interaction = keyword_data.nsmallest(15, 'Cost per Interaction')
bottom15_keywords_cost_per_interaction
```

```python
# change ' --' in interaction rate column to 0%
interaction_rate_list=keyword_data['Interaction rate'].to_list()
for i, x in enumerate(interaction_rate_list):
  if x ==' --':
    interaction_rate_list[i] = '0.00%'

# remove the percent sign from the interaction rate and change to decimals
interaction_rate_remove_percent_sign_list=[]
for i in interaction_rate_list:
  i = i[:-1]
  interaction_rate_remove_percent_sign_list.append(float(i))

interaction_rate_float_list = [i / 100 for i in interaction_rate_remove_percent_sign_list]

keyword_data['Interaction rate'] = interaction_rate_float_list
```

```python
# show the top 15 keywords based on Interaction Rate
top15_keywords_interaction_rate = keyword_data.nlargest(15, 'Interaction rate')                          
top15_keywords_interaction_rate
```

```python
# find the keywords which have lower average cost and higher interaction rate
low_cost_per_interatcion_list=bottom15_keywords_cost_per_interaction.index.to_list()
high_interaction_rate_list=top15_keywords_interaction_rate.index.to_list()

common_keywords = list(set.intersection(*map(set, [low_cost_per_interatcion_list, high_interaction_rate_list])))
if not common_keywords:
  print("No common keywords between lower average cost and higher interaction rate")
else:
  print(common_keywords)
```

**Find out Conversion, Cost per Conversion and Conversion Rate of ad_schedule_day_data**

```python
#Reformat ad_schedule_day_data dataframe
ad_schedule_day_data.columns = ad_schedule_day_data.iloc[1]
ad_schedule_day_data = ad_schedule_day_data[2:9]
ad_schedule_day_data
```

```python
ad_schedule_day_data.set_index('Day of the week', inplace=True)
ad_schedule_day_data
```

```python
#Convert the date in the Form Submitted column to datetime format
from datetime import datetime
weekdays=[]
for i in hubspot_crm_data['Form Submitted']:
  weekday = datetime.strptime(i,"%m/%d/%Y %H:%M")
  weekdays.append(weekday)
```

```python
#Add weekday column to the hubspot dataframe
#Each ID in the HubSpot table is a conversion
hubspot_crm_data['Form Submitted']=weekdays
hubspot_crm_data['Weekday'] = hubspot_crm_data['Form Submitted'].dt.strftime('%A')
hubspot_crm_data
```

```python
weekday_list = []
for day in hubspot_crm_data['Weekday']:
  weekday_list.append(day)
#weekday_list
week_day_occurance = [weekday_list.count('Saturday'),weekday_list.count('Sunday'),weekday_list.count('Friday'),weekday_list.count('Wednesday'),weekday_list.count('Thursday'),weekday_list.count('Tuesday'),weekday_list.count('Monday')]

#add Day of the Week's Conversion data into dataframe
ad_schedule_day_data['Total_Day_Conversions'] = week_day_occurance
ad_schedule_day_data
```

```python
#add Cost per Conversion and Conversion Rate into dataframe

day_int_cost_list=[]
day_cost_list=ad_schedule_day_data['Cost'].to_list()

for c in day_cost_list:
  day_int_cost_list.append(float(c))

ad_schedule_day_data['Cost']= day_int_cost_list


day_int_interaction_list=[]
day_interaction_list=ad_schedule_day_data['Interactions'].to_list()

for i in day_interaction_list:
  day_int_interaction_list.append(float(i))

ad_schedule_day_data['Interactions']= day_int_interaction_list


ad_schedule_day_data['Cost per Conversion'] = ad_schedule_day_data['Cost']/ad_schedule_day_data['Total_Day_Conversions']
ad_schedule_day_data['Conversion Rate'] = ad_schedule_day_data['Total_Day_Conversions']/ad_schedule_day_data['Interactions'] * 100
ad_schedule_day_data
```

**Line Chart of Daily Conversion Rate**

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
plt.plot('Conversion Rate', data= ad_schedule_day_data)
plt.legend()
```

**Find out Conversion, Cost per Conversion and Conversion Rate of ad_schedule_hour_data**

```python
#importing hourly data
ad_schedule_hour_data = pd.read_csv('/content/drive/Shareddrives/Python Project/Data/New_Data/Ad schedule hour report (1).csv')

#changing column names and trimming data
ad_schedule_hour_data.columns = ad_schedule_hour_data.iloc[1]
ad_schedule_hour_data=ad_schedule_hour_data.iloc[2:26]

#Changing hours column values from stings to ints
table_hours = ad_schedule_hour_data['Hour of the day'].to_list()

int_table_hours = []
for t in table_hours:
  int_table_hours.append(int(t))

ad_schedule_hour_data['Hour of the day']= int_table_hours

#Sorting the table by hour from 0-23
ad_schedule_hour_data.sort_values(by=['Hour of the day'], inplace=True, ascending=True)

#taking client data and extracting what hour each form was submitted
hubspot_crm_data['Hour_sent'] = hubspot_crm_data['Form Submitted'].dt.strftime('%H')

#List that contains hour submitted of every form
hour_list = []
for hour in hubspot_crm_data['Hour_sent']:
  hour_list.append(hour)

#Converting hout_list from string to int
int_hour_list = []
for h in hour_list:
  int_hour_list.append(int(h))

#counting how many instances of each hour are in the list
x = 0
hour_occurance = []
while x < 24:
  val = int_hour_list.count(x)
  hour_occurance.append(val)
  x += 1

#Creating a new column for total coversion per hour
ad_schedule_hour_data['Total_Hour_Conversions']= hour_occurance

#Changing hour of the day to index column
ad_schedule_hour_data.set_index('Hour of the day', inplace=True)
ad_schedule_hour_data
```

```python
#Changing cost column values from stings to ints
cost_list=ad_schedule_hour_data['Cost'].to_list()
int_cost_list=[]

for c in cost_list:
  int_cost_list.append(float(c))

ad_schedule_hour_data['Cost']= int_cost_list

#Changing interactoin column values from stings to ints
inter_list=ad_schedule_hour_data['Interactions'].to_list()
int_inter_list=[]

for i in inter_list:
  int_inter_list.append(float(i))

ad_schedule_hour_data['Interactions']= int_inter_list

#Creating cost per conversion and conversion rate columns
ad_schedule_hour_data['Cost per Conversion'] = ad_schedule_hour_data['Cost']/ad_schedule_hour_data['Total_Hour_Conversions']
ad_schedule_hour_data['Conversion Rate'] = ad_schedule_hour_data['Total_Hour_Conversions']/ad_schedule_hour_data['Interactions'] * 100
ad_schedule_hour_data

```

**Line Chart of Hourly Conversion Rate**

```python
#Line graph of coversion rates for each hour
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
plt.plot('Conversion Rate', data= ad_schedule_hour_data)
plt.legend()

#based off the data, our ads have the highest conversion rates from 11AM-9PM, peaking at 3PM (15% of all people who click out ads from 3-4 pm submit a form)
```
